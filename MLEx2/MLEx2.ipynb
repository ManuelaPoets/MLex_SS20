```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math
import random

theta=[random.uniform(-0.01, 0.01), random.uniform(-0.01, 0.01), random.uniform(-0.01, 0.01)]
#print(theta)
alpha=0.1

data= np.genfromtxt('data.txt', delimiter=' ')
#print(data)

pos= data[data[:,2] ==1]
neg= data[data[:,2] ==0]

x = [-3,3]
x2 = [(theta[0]+theta[1]*x[0])*(-1/theta[2]),(theta[0]+theta[1]*x[1])*(-1/theta[2]) ]
#print(data[:,0])
#print(npX)
plt.plot(x, x2, color ='blue', label='random classification') 


def g(theta, x):
    return 1/(1+np.exp(-z(theta,x)))

def z(theta,x):
    z = theta[0] + theta[1]*x[1] + theta[2]*x[2]
    return z

for iteration in range(400):
    for i in range(len(data)):
        #points
        x = [1, data[i][0], data[i][1]]
        #class
        y = data[i][2]
        
        for j in range(len(theta)):
            #print(g(theta, x))
            #print(x[j])
            theta[j] = theta[j] + alpha*(y-g(theta, x))*x[j]
            
            
plt.scatter(pos[:,0], pos[:,1],marker="o", color="green")
plt.scatter(neg[:,0], neg[:,1],marker="x", color="red")


x = [-3,3]
x2 = [(theta[0]+theta[1]*x[0])*(-1/theta[2]),(theta[0]+theta[1]*x[1])*(-1/theta[2]) ]
#print(x2)
plt.plot(x, x2, color="black", label='final classification')
plt.legend(framealpha=1, frameon=True)
plt.show()

```
